```python
def conv(value):
    if value == 1:
        return "yes"
    else :
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
        
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers using churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby("Churn").agg({'Churn':'count'})
colors =['#F5D627','#F52727']
plt.pie(gb['Churn'],labels =gb.index,autopct ='%1.2f%%',colors =colors)
plt.title("Percentage of Churned Cusotmers",fontsize =17)
plt.show()

```

 From the above piechart we can conclude that 26.54% of the customers have churned out .Now lets explore the reason


```python
plt.figure(figsize=(4,4))
sns.countplot(x ="gender",data =df,hue ="Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x ="SeniorCitizen",data =df)
ax.bar_label(ax.containers[0])
plt.title("Count of Senior Cetizen")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crosstab counts
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Convert to %
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar
plt.figure(figsize=(4,4))
ax = ct_pct.plot(kind="bar", stacked=True)

# Title & labels
plt.title("Churn by Senior Citizen (Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage Distribution")

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.legend(title="Churn")
plt.show()
```

comparative a greater percentage of senior citizen category have churend out

```python
# Import required libraries
# import matplotlib.pyplot as plt
# import seaborn as sns

# List of categorical columns you want to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create a big figure (canvas)
plt.figure(figsize=(15, 10))

# Loop through each column
for i, col in enumerate(columns):
    
    # Create subplot grid (3 rows, 3 columns)
    plt.subplot(3, 3, i + 1)
    
    # Draw countplot for current column
    sns.countplot(x=col, data=df,hue ="Churn")
    
    # Add title
    plt.title(f"Count of {col}")
    
    # Rotate x-axis labels for readability
    plt.xticks(rotation=45)

# Adjust spacing between plots
plt.tight_layout()

# Show all plots
plt.show()
```
